```python
#question no 3

n = int(input("Enter a positive integer (N): "))
total_sum = 0
current_number = 1
while current_number <= n:
    total_sum += current_number  
    current_number += 1         
print(f"The sum of numbers from 1 to {n} is: {total_sum}")
```

```python
#question no 4

num = int(input("Enter a number to see its multiplication table:"))
i = 1
while i <= 10:
    result = num * i
    print(f"{num} x {i} = {result}")
    i += 1
```

```python
#question no 7

num = int(input("Enter a number to count its digits: "))
if num == 0:
    count = 1
else:
    count = 0
    temp_num = abs(num)
    
    while temp_num > 0:
        temp_num = temp_num // 10  
        count += 1                
print(f"The number of digits is: {count}")
```

```python
#question no 8

total = 0  

while True:
    num = int(input("Enter a number to add (or 0 to stop): "))
    
    if num == 0:
        break
    
    total += num  
    print(f"Current subtotal: {total}") 

print("--- Final Result ---")
print(f"The total sum of all numbers entered is: {total}")
```

```python
#question no 9

num = 101
while num % 7 != 0:
    num += 1
print(f"The first number greater than 100 divisible by 7 is: {num}")
```

```python
#question no 10
secret_number = 25
guess = None
attempts = 0

print(" Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 50.")

while guess != secret_number:
    guess = int(input("Enter your guess: "))
    attempts += 1  
    
    # 3. Give hints
    if guess < secret_number:
        print("Too low! Try again. ")
    elif guess > secret_number:
        print("Too high! Try again. ")
    else:
        
        print(f"🎉 Correct! You found it in {attempts} attempts.")
```
